```python
from string import ascii_lowercase
```

```python
alphabet = [c for c in ascii_lowercase]
blanc_term = "<blanc>"
terms = [blanc_term] + alphabet
```

```python
idx2char = dict(enumerate(terms))
char2idx = {c: i for i,c in enumerate(terms)}
```

```python
char2idx
```

```python
import torch
from torch.nn.functional import one_hot
```

```python
mean_sign_expo = 4
mean_pause = 2

blanc_idx = char2idx[blanc_term]

target_seq = ['a','c','b']
target_seq_idx = [char2idx[c] for c in target_seq]
target_seq_idx
```

```python
target_tensor = torch.tensor([target_seq_idx])
target_tensor
```

```python
stream_seq = [[i]*3+[blanc_idx]*2 for i in target_seq_idx]
stream_seq = [x for xs in stream_seq for x in xs]
stream_seq
```

```python
stream_tensor = one_hot(torch.tensor(stream_seq), num_classes=len(char2idx))
stream_tensor = torch.unsqueeze(stream_tensor, 1) # batch dim is 1
stream_tensor
```

```python
out_tensor = (stream_tensor + torch.randn(stream_tensor.shape)*0.01).log_softmax(2)

out_tensor.shape
```

```python
input_lengths = torch.tensor([[15]])
target_lengths = torch.tensor([[3]])
```

```python
ctc_loss = torch.nn.CTCLoss()
ctc_loss(out_tensor, target_tensor, input_lengths, target_lengths)
```

```python
target2 = ["a", "b"]
target2_tensor = torch.tensor([[char2idx[c] for c in target2]])
target2_lengths = torch.tensor([[2]])

ctc_loss(out_tensor, target2_tensor, input_lengths, target2_lengths)
```

```python
target3 = ["a", "c"]
target3_tensor = torch.tensor([[char2idx[c] for c in target3]])
target3_lengths = torch.tensor([[2]])

ctc_loss(out_tensor, target3_tensor, input_lengths, target3_lengths)
```
